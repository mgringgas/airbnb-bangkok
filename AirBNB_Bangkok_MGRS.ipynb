## Data Cleaning

Data mentah sering mengandung masalah seperti missing values, duplikat, format tidak konsisten, atau outlier. Data cleaning adalah proses memperbaiki masalah ini agar data akurat, konsisten, dan siap dianalisis, sehingga hasil analisis lebih valid dan dapat diandalkan.

### Langkah-Langkah Data Cleaning

Berdasarkan temuan dari tahap Data Understanding, berikut adalah langkah-langkah pembersihan data yang perlu kita lakukan pada dataset Airbnb Bangkok:

1.  **Menangani Nilai yang Hilang (Missing Values):**
    *   Kolom `name` dan `host_name` memiliki jumlah nilai hilang yang sedikit. Kita bisa mempertimbangkan untuk menghapus baris yang memiliki nilai hilang pada kolom ini atau melakukan imputasi sederhana jika relevan.
    *   Kolom `last_review` dan `reviews_per_month` memiliki jumlah nilai hilang yang signifikan (5790 baris). Karena nilai hilang ini terkait dengan listing yang belum memiliki ulasan (`number_of_reviews` adalah 0), kita bisa mempertimbangkan:
        *   Mengisi nilai `reviews_per_month` yang hilang dengan 0.
        *   Menentukan strategi penanganan untuk `last_review` yang hilang; mungkin tidak relevan untuk analisis temporal jika tidak ada ulasan, atau bisa diisi dengan nilai placeholder jika diperlukan.

```python
# Menangani nilai yang hilang

# Mengisi nilai yang hilang pada 'reviews_per_month' dengan 0
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

# Menghapus baris dengan nilai yang hilang pada 'name' dan 'host_name'
df.dropna(subset=['name', 'host_name'], inplace=True)

# Memeriksa kembali jumlah nilai yang hilang setelah penanganan
print("Jumlah Nilai yang Hilang setelah Penanganan:")
print(df.isnull().sum())
```

2.  **Mengatasi Outlier dan Nilai Tidak Valid pada Kolom `price`:**
    *   Memeriksa listing dengan `price` 0 dan memutuskan apakah akan menghapusnya atau menganggapnya sebagai data tidak valid.
    *   Mengidentifikasi dan menangani outlier pada kolom `price` yang memiliki nilai sangat tinggi. Metode penanganan bisa dengan menghapus outlier, capping (membatasi nilai maksimum), atau transformasi data, tergantung pada distribusi data dan tujuan analisis.



```python
# Mengatasi Outlier dan Nilai Tidak Valid pada Kolom 'price'
# Menggunakan capping pada persentil ke-99

# Menghapus baris dengan price 0
if 0 in df['price'].unique():
    print(f"Jumlah baris sebelum menghapus price = 0: {len(df)}")
    df = df[df['price'] > 0]
    print(f"Jumlah baris setelah menghapus price = 0: {len(df)}")

# Menghitung nilai pada persentil ke-99 dari kolom 'price'
upper_threshold = df['price'].quantile(0.99)

print(f"\nNilai price pada persentil ke-99: {upper_threshold}")

# Melakukan capping pada nilai price yang melebihi upper_threshold
# Nilai price yang lebih besar dari upper_threshold akan diganti dengan upper_threshold
df['price'] = df['price'].clip(upper=upper_threshold)

print("\nStatistik deskriptif kolom price setelah penanganan outlier (capping):")
display(df['price'].describe())

print(f"\nJumlah baris setelah penanganan outlier price: {len(df)}")
```

3.  **Konversi Tipe Data:**
    *   Mengkonversi kolom `last_review` dari tipe `object` menjadi tipe data `datetime` agar dapat digunakan untuk analisis berbasis waktu.


```python
# Konversi tipe data kolom 'last_review' menjadi datetime
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Memeriksa kembali tipe data kolom 'last_review'
print("Tipe data kolom 'last_review' setelah konversi:")
print(df['last_review'].dtypes)

# Memeriksa kembali jumlah nilai yang hilang pada 'last_review'
print("\nJumlah nilai yang hilang pada 'last_review' setelah konversi:")
print(df['last_review'].isnull().sum())

# Tampilkan beberapa baris kolom 'last_review' untuk verifikasi
print("\nContoh data kolom 'last_review':")
display(df['last_review'].head())
```

4.  **Menghapus Kolom yang Tidak Diperlukan:**
    *   Kolom `Unnamed: 0` tampaknya merupakan indeks data yang tidak relevan untuk analisis, sehingga dapat dihapus.
    *   Memeriksa kolom lain seperti `id` dan `host_id` apakah perlu dihapus atau dipertahankan tergantung pada apakah kita akan menggunakannya untuk identifikasi unik atau analisis terkait host/listing tertentu. Untuk sebagian besar analisis agregat, kolom ID mungkin tidak diperlukan.

```python
# Menghapus Kolom yang Tidak Diperlukan

# Menghapus kolom 'Unnamed: 0'
df.drop('Unnamed: 0', axis=1, inplace=True)

# Memeriksa kembali nama-nama kolom setelah penghapusan
print("\nNama-nama kolom setelah penghapusan:")
df.columns
```

### Pertanyaan 1: Karakteristik dan Popularitas Listing

Untuk menjawab pertanyaan pertama mengenai karakteristik dan popularitas listing, kita akan melakukan analisis berikut secara bertahap:

#### Langkah 3.1: Hubungan Tipe Kamar dan Popularitas

**Analisis:** Mengelompokkan data berdasarkan `room_type` dan menghitung rata-rata `number_of_reviews` untuk setiap tipe kamar.

**Tujuan:** Untuk melihat tipe kamar mana yang cenderung paling banyak mendapatkan ulasan. Jumlah ulasan yang tinggi dianggap sebagai indikator popularitas listing.

```python
# Kelompokkan data berdasarkan 'room_type' dan hitung rata-rata 'number_of_reviews'
popularity_by_room_type = df.groupby('room_type')['number_of_reviews'].mean().sort_values(ascending=False)

print("Rata-rata Jumlah Ulasan per Tipe Kamar:")
display(popularity_by_room_type)

# Visualisasi hasil
plt.figure(figsize=(8, 6))
sns.barplot(x=popularity_by_room_type.index, y=popularity_by_room_type.values)
plt.title('Rata-rata Jumlah Ulasan per Tipe Kamar')
plt.xlabel('Tipe Kamar')
plt.ylabel('Rata-rata Jumlah Ulasan')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

#### Insight dari Langkah 3.1: Hubungan Tipe Kamar dan Popularitas (berdasarkan Rata-rata Ulasan)

Berdasarkan analisis dan visualisasi rata-rata jumlah ulasan per tipe kamar:

*   Listing dengan tipe **Entire home/apt** memiliki rata-rata ulasan tertinggi (sekitar 22.21 ulasan per listing).
*   Diikuti oleh **Hotel room** dengan rata-rata sekitar 19.60 ulasan per listing.
*   Sementara itu, **Private room** dan **Shared room** memiliki rata-rata ulasan yang lebih rendah (sekitar 8.80 dan 5.13 ulasan per listing).

Ini mengindikasikan bahwa, secara rata-rata, listing yang menawarkan seluruh properti (`Entire home/apt`) atau merupakan bagian dari hotel (`Hotel room`) cenderung mendapatkan lebih banyak ulasan dibandingkan dengan kamar pribadi atau kamar bersama. Hal ini bisa diinterpretasikan bahwa kedua tipe kamar pertama tersebut lebih populer atau mungkin memiliki tingkat hunian yang lebih tinggi yang menghasilkan lebih banyak interaksi dan ulasan dari tamu.

**Analisis:** Mengelompokkan data berdasarkan `neighbourhood` dan menghitung jumlah total ulasan (`number_of_reviews`) untuk setiap lingkungan. Visualisasikan dalam bar chart.

Berdasarkan analisis dan visualisasi total jumlah ulasan per lingkungan (`neighbourhood`), didapatkan beberapa *insight* utama mengenai popularitas geografi listing Airbnb di Bangkok:

*   Distribusi total ulasan menunjukkan bahwa popularitas listing Airbnb tidak merata di seluruh Bangkok, melainkan terkonsentrasi di wilayah-wilayah tertentu. Wilayah-wilayah yang paling populer ini seringkali merupakan pusat bisnis, hiburan, perbelanjaan, atau memiliki akses transportasi yang baik.

Pada bagian ini, kita akan mendalami faktor-faktor yang memengaruhi harga listing Airbnb di Bangkok dan melihat bagaimana harga bervariasi di seluruh pasar. Analisis ini akan membantu memahami dinamika penetapan harga di berbagai segmen dan wilayah.

#### Langkah 3.4: Hubungan Harga dan Tipe Kamar

**Analisis:** Membuat boxplot untuk membandingkan sebaran `price` dari setiap `room_type` secara visual.

**Tujuan:** Melihat tipe kamar mana yang memiliki harga rata-rata tertinggi dan terendah, serta memahami variasi harga di dalam setiap tipe kamar.

```python
# Buat boxplot harga berdasarkan tipe kamar
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='room_type', y='price')
plt.title('Sebaran Harga (Price) per Tipe Kamar')
plt.xlabel('Tipe Kamar')
plt.ylabel('Harga (Price)')
plt.xticks(rotation=45, ha='right')
plt.gca().yaxis.set_major_formatter(mticker.ScalarFormatter(useOffset=False, useMathText=False))
plt.ticklabel_format(style='plain', axis='y', useOffset=False)
plt.tight_layout()
plt.show()
```

#### Insight dari Langkah 3.4: Hubungan Harga dan Tipe Kamar

### Hipotesis 1: Listing dengan tipe "Entire home/apt" memiliki harga rata-rata yang signifikan lebih tinggi dibandingkan tipe kamar lainnya.

**Latar Belakang:** Dari EDA (Langkah 3.4), kita melihat bahwa "Entire home/apt" cenderung memiliki median harga yang lebih tinggi. Hipotesis ini akan menguji apakah perbedaan harga ini signifikan secara statistik.

**Metode Uji:** Kita bisa menggunakan uji statistik seperti uji ANOVA (Analysis of Variance) jika membandingkan lebih dari dua kelompok, atau uji t independen jika membandingkan dua kelompok. Dalam kasus ini, kita akan membandingkan "Entire home/apt" dengan tipe kamar lainnya, sehingga uji ANOVA lebih tepat.

**Langkah-Langkah Pengujian:**
1. Ambil sampel data harga untuk setiap tipe kamar.
2. Lakukan uji ANOVA untuk membandingkan rata-rata harga antar tipe kamar.
3. Interpretasikan hasil uji (nilai p dan statistik F) untuk menentukan apakah ada perbedaan rata-rata harga yang signifikan.

```python
import scipy.stats as stats

# Ambil data harga untuk setiap tipe kamar
entire_home_price = df[df['room_type'] == 'Entire home/apt']['price']
hotel_room_price = df[df['room_type'] == 'Hotel room']['price']
private_room_price = df[df['room_type'] == 'Private room']['price']
shared_room_price = df[df['room_type'] == 'Shared room']['price']

# Uji ANOVA
f_statistic, p_value = stats.f_oneway(entire_home_price, hotel_room_price, private_room_price, shared_room_price)

print(f"Hasil Uji ANOVA untuk Hipotesis 1:")
print(f"Statistik F: {f_statistic:.4f}")
print(f"Nilai p: {p_value:.4f}")

# Interpretasi hasil
alpha = 0.05
print(f"\nTingkat signifikansi (alpha): {alpha}")

if p_value < alpha:
    print("\nKesimpulan: Tolak Hipotesis Nol (H0).")
    print("Ada perbedaan rata-rata harga yang signifikan secara statistik antar tipe kamar.")
else:
    print("\nKesimpulan: Gagal Menolak Hipotesis Nol (H0).")
    print("Tidak ada perbedaan rata-rata harga yang signifikan secara statistik antar tipe kamar.")
```

#### Kesimpulan Hipotesis 1

Berdasarkan hasil uji ANOVA, diperoleh nilai p (p-value) sebesar 0.0000, yang lebih kecil dari tingkat signifikansi (alpha) 0.05. Dengan demikian, kita menolak Hipotesis Nol (H0).

Artinya, terdapat perbedaan rata-rata harga yang signifikan secara statistik antar tipe kamar di Airbnb Bangkok. Secara khusus, listing dengan tipe "Entire home/apt" cenderung memiliki harga rata-rata yang lebih tinggi dibandingkan tipe kamar lainnya, dan perbedaan ini signifikan secara statistik.

### Hipotesis 2: Lingkungan (Neighbourhood) dengan total jumlah ulasan tertinggi memiliki rata-rata harga yang signifikan lebih tinggi.

**Latar Belakang:** Dari EDA (Langkah 3.3 dan 3.5), kita melihat bahwa beberapa lingkungan sangat populer (total ulasan tinggi) dan ada variasi harga antar lingkungan. Hipotesis ini akan menguji apakah popularitas suatu lingkungan berkorelasi positif dengan rata-rata harganya.

**Metode Uji:** Kita bisa menguji korelasi antara total ulasan per lingkungan dan rata-rata harga per lingkungan, atau membandingkan rata-rata harga antara kelompok lingkungan (misalnya, 10 lingkungan terpopuler vs. 10 lingkungan lainnya).

**Langkah-Langkah Pengujian:**
1. Hitung total ulasan dan rata-rata harga per lingkungan.
2. Lakukan analisis korelasi atau uji statistik perbandingan rata-rata (misalnya, uji t atau ANOVA tergantung pengelompokan) antara popularitas dan harga lingkungan.
3. Interpretasikan hasilnya.

```python
# Hitung total ulasan per lingkungan
popularity_by_neighbourhood = df.groupby('neighbourhood')['number_of_reviews'].sum()

# Hitung rata-rata harga per lingkungan
average_price_by_neighbourhood = df.groupby('neighbourhood')['price'].mean()

# Gabungkan kedua Series berdasarkan index (neighbourhood)
neighbourhood_analysis = pd.merge(popularity_by_neighbourhood, average_price_by_neighbourhood, left_index=True, right_index=True)
neighbourhood_analysis.rename(columns={'number_of_reviews': 'total_reviews', 'price': 'average_price'}, inplace=True)

# Lakukan analisis korelasi
correlation = neighbourhood_analysis['total_reviews'].corr(neighbourhood_analysis['average_price'])

print(f"Hasil Analisis Korelasi antara Total Ulasan dan Rata-rata Harga per Lingkungan:")
print(f"Koefisien Korelasi Pearson: {correlation:.4f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Hasil: Tolak H0 → Ada hubungan signifikan antara total ulasan dan rata-rata harga.")
else:
    print("Hasil: Gagal menolak H0 → Tidak ada hubungan signifikan antara total ulasan dan rata-rata harga.")

# Visualisasi scatter plot untuk melihat hubungan
plt.figure(figsize=(10, 6))
sns.scatterplot(data=neighbourhood_analysis, x='total_reviews', y='average_price', alpha=0.7)
plt.title('Hubungan Total Ulasan dan Rata-rata Harga per Lingkungan')
plt.xlabel('Total Jumlah Ulasan per Lingkungan')
plt.ylabel('Rata-rata Harga per Lingkungan')
plt.show()
```

```python
from scipy.stats import pearsonr

r, p_value = pearsonr(neighbourhood_analysis['total_reviews'], neighbourhood_analysis['average_price'])

print(f"Koefisien Korelasi Pearson: {r:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Hasil: Tolak H0 → Ada hubungan signifikan antara total ulasan dan rata-rata harga.")
else:
    print("Hasil: Gagal menolak H0 → Tidak ada hubungan signifikan antara total ulasan dan rata-rata harga.")
```

#### Kesimpulan Hipotesis 2

Berdasarkan hasil uji korelasi Pearson, diperoleh koefisien korelasi sebesar 0.2596 dengan nilai p-value 0.0687. Nilai korelasi ini menunjukkan adanya hubungan positif yang lemah antara total jumlah ulasan dan rata-rata harga per lingkungan. Namun, karena p-value lebih besar dari taraf signifikansi 0.05, maka gagal menolak H0. Dengan demikian, tidak terdapat bukti yang cukup untuk menyatakan adanya hubungan yang signifikan antara popularitas lingkungan (berdasarkan jumlah ulasan) dengan rata-rata harga. Hal ini berarti Hipotesis 2 tidak didukung oleh data.

### Hipotesis 3: Host dengan jumlah listing yang banyak ("Multiple Listing Host") memiliki tingkat aktivitas listing (reviews per month) yang signifikan lebih tinggi dibandingkan host dengan satu listing ("Single Listing Host").

**Latar Belakang:** Dari EDA (Langkah 3.6), kita melihat bahwa "Multiple Listing Host" memiliki rata-rata jumlah ulasan total yang lebih tinggi. Hipotesis ini menguji apakah tren ini juga berlaku untuk aktivitas ulasan bulanan, yang lebih relevan untuk mengukur popularitas saat ini.

**Metode Uji:** Kita bisa menggunakan uji statistik perbandingan rata-rata, seperti uji t independen, untuk membandingkan rata-rata `reviews_per_month` antara dua kelompok host (`Single Listing Host` dan `Multiple Listing Host`).

**Langkah-Langkah Pengujian:**
1. Bagi data berdasarkan `host_type`.
2. Lakukan uji t independen pada kolom `reviews_per_month` antara kedua kelompok.
3. Interpretasikan hasil uji t.

```python
# Bagi data berdasarkan host_type
single_listing_reviews = df[df['host_type'] == 'Single Listing Host']['reviews_per_month']
multiple_listing_reviews = df[df['host_type'] == 'Multiple Listing Host']['reviews_per_month']

# Lakukan uji t independen
# Asumsi: independent samples (dua kelompok host berbeda)
# Equal variance: bisa diuji dengan Levene's test, atau gunakan parameter equal_var=False jika tidak yakin
# Untuk saat ini, kita asumsikan equal variance (default) atau bisa diubah jika perlu
ttest_statistic, p_value = stats.ttest_ind(single_listing_reviews, multiple_listing_reviews, equal_var=False) # Menggunakan Welch's t-test (equal_var=False) yang lebih robust jika variansi tidak sama

print(f"Hasil Uji t Independen untuk Hipotesis 3:")
print(f"Statistik t: {ttest_statistic:.4f}")
print(f"Nilai p: {p_value:.4f}")

# Interpretasi hasil
alpha = 0.05
print(f"\nTingkat signifikansi (alpha): {alpha}")

if p_value < alpha:
    print("\nKesimpulan: Tolak Hipotesis Nol (H0).")
    print("Ada perbedaan rata-rata reviews_per_month yang signifikan secara statistik antara Single Listing Host dan Multiple Listing Host.")
    # Cek grup mana yang lebih tinggi rata-ratanya
    if single_listing_reviews.mean() > multiple_listing_reviews.mean():
        print(f"Rata-rata reviews_per_month Single Listing Host ({single_listing_reviews.mean():.4f}) lebih tinggi dari Multiple Listing Host ({multiple_listing_reviews.mean():.4f}).")
    else:
         print(f"Rata-rata reviews_per_month Multiple Listing Host ({multiple_listing_reviews.mean():.4f}) lebih tinggi dari Single Listing Host ({single_listing_reviews.mean():.4f}).")

else:
    print("\nKesimpulan: Gagal Menolak Hipotesis Nol (H0).")
    print("Tidak ada perbedaan rata-rata reviews_per_month yang signifikan secara statistik antara Single Listing Host dan Multiple Listing Host.")

# Hipotesis Nol (H0): Rata-rata reviews_per_month untuk Single Listing Host sama dengan Multiple Listing Host.
# Hipotesis Alternatif (H1): Rata-rata reviews_per_month untuk Single Listing Host tidak sama dengan Multiple Listing Host.
```

#### Kesimpulan Hipotesis 3

Berdasarkan hasil uji t independen, diperoleh nilai p (p-value) sebesar 0.0000, yang lebih kecil dari tingkat signifikansi (alpha) 0.05.

Dengan demikian, kita menolak Hipotesis Nol (H0).

Artinya, terdapat perbedaan rata-rata tingkat aktivitas listing (reviews per month) yang signifikan secara statistik antara listing yang dikelola oleh Single Listing Host dan Multiple Listing Host. Secara khusus, rata-rata reviews per month untuk Multiple Listing Host (0.6193) lebih tinggi dibandingkan dengan Single Listing Host (0.2649).

### Hipotesis 4: Lingkungan (Neighbourhood) dengan rata-rata ketersediaan (availability_365) terendah memiliki rata-rata jumlah ulasan (number_of_reviews) yang signifikan lebih tinggi.

**Latar Belakang:** Dari EDA (Langkah 3.10), kita melihat bahwa beberapa lingkungan memiliki rata-rata ketersediaan listing yang rendah, yang bisa mengindikasikan permintaan tinggi. Hipotesis ini menguji apakah ketersediaan rendah berkorelasi dengan popularitas yang tinggi (berdasarkan jumlah ulasan).

**Metode Uji:** Kita bisa menguji korelasi antara rata-rata ketersediaan per lingkungan dan total ulasan per lingkungan, atau membandingkan total ulasan antara kelompok lingkungan (misalnya, 10 lingkungan dengan ketersediaan terendah vs. 10 lingkungan lainnya).

**Langkah-Langkah Pengujian:**
1. Hitung rata-rata ketersediaan dan total ulasan per lingkungan.
2. Lakukan analisis korelasi atau uji statistik perbandingan rata-rata antara ketersediaan dan popularitas lingkungan.
3. Interpretasikan hasilnya.

```python
# Hitung rata-rata ketersediaan per lingkungan
average_availability_by_neighbourhood = df.groupby('neighbourhood')['availability_365'].mean()

# Hitung total ulasan per lingkungan
total_reviews_by_neighbourhood = df.groupby('neighbourhood')['number_of_reviews'].sum()

# Gabungkan kedua Series berdasarkan index (neighbourhood)
neighbourhood_availability_reviews = pd.merge(average_availability_by_neighbourhood, total_reviews_by_neighbourhood, left_index=True, right_index=True)
neighbourhood_availability_reviews.rename(columns={'availability_365': 'average_availability', 'number_of_reviews': 'total_reviews'}, inplace=True)

# Lakukan analisis korelasi
r, p_value = pearsonr(neighbourhood_availability_reviews['average_availability'], neighbourhood_availability_reviews['total_reviews'])

print(f"Hasil Analisis Korelasi antara Rata-rata Ketersediaan dan Total Ulasan per Lingkungan:")
print(f"Koefisien Korelasi Pearson: {r:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
print(f"\nTingkat signifikansi (alpha): {alpha}")

if p_value < alpha:
    print("\nKesimpulan: Tolak Hipotesis Nol (H0).")
    if r < 0:
        print("Ada hubungan negatif yang signifikan antara rata-rata ketersediaan dan total ulasan per lingkungan.")
        print("Lingkungan dengan ketersediaan lebih rendah cenderung memiliki total ulasan lebih tinggi.")
    else:
        print("Ada hubungan positif yang signifikan antara rata-rata ketersediaan dan total ulasan per lingkungan.")
        print("Lingkungan dengan ketersediaan lebih tinggi cenderung memiliki total ulasan lebih tinggi.")
else:
    print("\nKesimpulan: Gagal Menolak Hipotesis Nol (H0).")
    print("Tidak ada hubungan signifikan antara rata-rata ketersediaan dan total ulasan per lingkungan.")

# Visualisasi scatter plot untuk melihat hubungan
plt.figure(figsize=(10, 6))
sns.scatterplot(data=neighbourhood_availability_reviews, x='average_availability', y='total_reviews', alpha=0.7)
plt.title('Hubungan Rata-rata Ketersediaan dan Total Ulasan per Lingkungan')
plt.xlabel('Rata-rata Jumlah Hari Tersedia dalam Setahun per Lingkungan')
plt.ylabel('Total Jumlah Ulasan per Lingkungan')
plt.show()
```

#### Kesimpulan Hipotesis 4

Berdasarkan hasil uji korelasi Pearson antara rata-rata ketersediaan (average_availability_by_neighbourhood) dan total jumlah ulasan (total_reviews_by_neighbourhood) per lingkungan, diperoleh koefisien korelasi sebesar -0.3839 dengan nilai p-value sebesar 0.0059.

Dengan tingkat signifikansi (alpha) 0.05, karena p-value < alpha, maka kita menolak Hipotesis Nol (H0).

Artinya, terdapat hubungan negatif yang signifikan antara rata-rata ketersediaan dan total ulasan per lingkungan. Hal ini berarti lingkungan dengan ketersediaan yang lebih rendah cenderung memiliki total ulasan yang lebih tinggi. Dengan demikian, hasil ini mendukung Hipotesis 4.
